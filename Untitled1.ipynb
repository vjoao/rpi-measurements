```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set()
```

```python
fedberry = pd.read_csv('arch-boot-wifi.csv')
```

```python
plt.figure(figsize=(14, 7))
plot = fedberry.set_index('time').power.plot()
plot.set_xlabel('Tempo (ms)')
plot.set_ylabel('Potência (W)')
```

```python
def energy (time, power):
    return power * (time / 1000 / 3600);

fedberry['delta'] = fedberry.time.diff().shift(-1)
fedberry['energy'] = fedberry.apply(lambda row: energy(row.delta, row.power), axis=1)
```

```python
plt.figure(figsize=(14, 7))
plot = fedberry.set_index('time').energy.cumsum().plot()
plot.set_ylabel('Energia consumida (Wh)')
plot.set_xlabel('Tempo (ms)')
```
