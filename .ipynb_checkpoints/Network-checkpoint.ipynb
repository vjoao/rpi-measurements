```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import seaborn as sns
sns.set()

plt.rc('axes', titlesize=16)

df = pd.read_csv('5mb-download-cable.csv')
df['network'] = 'Cabo'
df['mode'] = 'Download'

df2 = pd.read_csv('5mb-download-wifi.csv')
df2['network'] = 'Wifi'
df2['mode'] = 'Download'

df3 = pd.read_csv('5mb-upload-cable.csv')
df3['network'] = 'Cabo'
df3['mode'] = 'Upload'

df4 = pd.read_csv('5mb-upload-wifi.csv')
df4['network'] = 'Wifi'
df4['mode'] = 'Upload'

def energy (time, power):
    return power * (time / 1000 / 3600);

for d in [df, df2, df3, df4]:
    time = d.time.max()
    d['energy'] = d.apply(lambda row: energy(time, row.power), axis=1)
```

```python
master = pd.concat([df, df2, df3, df4])
```

```python
plt.figure(figsize=(14, 7))
plot = sns.barplot (x='network', y='energy', hue='mode', data=master)
plot.set_ylabel('Consumo (Wh)')
plot.set_title('Consumo das operações de rede (Arquivo 5MB)')

plot.set_xlabel('Conexão de rede')
plot.legend_.set_title('Modo')
```

```python
datasets = []

def energy (time, power):
    return power * (time / 1000 / 3600);

for algo in ['aes', 'aes256', '2fish']:
    for size in [1, 10, 25, 50, 100]:
        data = pd.read_csv('gpg-' + str(size) + 'mb-' + algo + '.csv')
        time = data.max().time
        
        data['size'] = size
        data['algo'] = algo
        data['energy'] = data.apply(lambda row: energy(time, row.power), axis=1)
        datasets.append(data)
```

```python
frame = pd.concat(datasets)
```

```python
frame.head()



```

```python
plt.figure(figsize=(14, 7))
plot = sns.barplot(x='algo', y='energy', hue='size', data=frame)
plot.legend_.set_title('Tamanho (MB)')
plot.set_xlabel('Algoritmo de criptografia')
plot.set_ylabel('Consumo de energia (Wh)')
plot.set_title('Consumo energético de criptografia')
```
