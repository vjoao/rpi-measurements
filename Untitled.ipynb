```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.linspace(0, 10, 1000)

df = pd.read_csv('fedberry-barebone-offline.csv')

df2 = pd.read_csv('fedberry-barebone-cable.csv')
df3 = pd.read_csv('fedberry-barebone-wifi.csv')

df['network'] = 'Não'
df2['network'] = 'Ethernet'
df3['network'] = 'Ethernet'
```

```python
df.mean()
```

```python
df2.mean()
```

```python
import seaborn as sns
sns.set()

plt.rc('axes', titlesize=16)

# plt.hist(df.power, bins=5, color=sns.desaturate("green", 1))
# sns.boxplot(df.power)
plt.figure(figsize=(14, 7))
plot = sns.kdeplot(df.merge(df2.merge(df3)).power, hue='network', shade=True)
plot.set_xlabel('Potência (W)')
plot.set_title('Fedberry - KDE de potência na inicialização')

# sns.distplot(df.power)
# sns.violinplot(df.current)
# sns.kdeplot(df.power, df.current, cmap=sns.cubehelix_palette(light=1, as_cmap=True))
```
